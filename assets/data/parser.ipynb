```python
import json
```

```python
with open('cedict_ts.u8', encoding='utf8') as file:
    text = file.read()
    lines = text.split('\n')
    dict_lines = list(lines)
```

```python
list_of_dicts = []
```

```python
dict_lines[5:31]
```

```python
# no need for pinyin
def parse_line(line):
    parsed = {}

    # handling empty lines
    if line == '':
        dict_lines.remove(line)
        return 0
    line = line.rstrip('/')
    line = line.split('/')
    if len(line) <= 1:
        return 0
    try:
        english = line[1]
        char_and_pinyin = line[0].split('[')
        characters = char_and_pinyin[0]
        characters = characters.split()
        traditional = characters[0]
        simplified = characters[1]

        # print('english (line1)', english)
        # print('char and py', char_and_pinyin)
        # print('trad (line[0][0])', traditional)
        # print('simp (line[0][1])', simplified)

        pinyin = char_and_pinyin[1]
        pinyin = pinyin.rstrip()
        pinyin = pinyin.rstrip("]")
        parsed['traditional'] = traditional
        parsed['simplified'] = simplified
        parsed['pinyin'] = pinyin
        parsed['english'] = english
        list_of_dicts.append(parsed)
    except:
        # print('error', e)
        print('erroneous line', line)


def remove_surnames():
    for x in range(len(list_of_dicts)-1, -1, -1):
        if "surname " in list_of_dicts[x]['english']:
            if list_of_dicts[x]['traditional'] == list_of_dicts[x+1]['traditional']:
                list_of_dicts.pop(x)
```

```python
def parse_dict():
    print("Parsing dictionary . . .")
    # first 31 lines are licensing stuff
    for line in dict_lines[32:]:
    # for line in dict_lines:
        parse_line(line)

    # remove entries for surnames from the data (optional):

    print("Removing Surnames . . .")
    remove_surnames()

    return list_of_dicts
```

```python
list_of_dicts = parse_dict()
list_of_dicts
```

```python
# send list of dicts to json
file_path = "cedict.json"
with open(file_path, "w", encoding="utf-8") as json_file:
    # json.dump(list_of_dicts, json_file)
    json.dump(list_of_dicts, json_file, ensure_ascii=False)
```
